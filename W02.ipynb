```python
import pandas as pd
import itertools as it
```

### Processing the data

```python
file = open("W02_data_groceries.csv")
s = file.read()
file.close()
```

```python
baskets = s.split("\n")
baskets = [x.split(",") for x in baskets]
#baskets
```

```python
SKUs = s.replace("\n",",")
SKUs = SKUs.split(",")
SKUs = list(set(SKUs))
SKUs.remove("")
SKUs.sort()
print('There are', len(SKUs), 'SKUs in total')
#print(SKUs)
```

```python
shelf1 = ['shopping bags','newspapers','long life bakery product','cling film/bags','flower (seeds)','pet care','photo/film','candles','dog food','rice','instant coffee','kitchen towels','light bulbs','preservation products','sound storage medium']
shelf2 = ['soda','bottled water','bottled beer','canned beer','fruit/vegetable juice','red/blush wine','white wine','liquor','sparkling wine','rum','brandy','cocoa drinks','prosecco','liqueur','whisky']
shelf3 = ['rolls/buns','sausage','pastry','brown bread','frankfurter','pork','beef','white bread','waffles','hamburger meat','zwieback','cereals','syrup','fish','honey']
shelf4 = ['napkins','hygiene articles','softener','cleaner','male cosmetics','abrasive cleaner','skin care','bathroom cleaner','decalcifier','hair spray','make up remover','toilet cleaner','baby cosmetics','kitchen utensil','baby food']
shelf5 = ['chocolate','salty snack','specialty chocolate','candy','specialty bar','chewing gum','cake bar','chocolate marshmallow','popcorn','nuts/prunes','artif. sweetener','snack products','cookware','cooking chocolate','pudding powder']
shelf6 = ['other vegetables','root vegetables','tropical fruit','citrus fruit','pip fruit','sauces','spices','ketchup','tea','canned fruit','potato products','organic sausage','ready soups','specialty vegetables','salad dressing']
shelf7 = ['whole milk','yogurt','butter','curd','frozen meals','spread cheese','frozen dessert','condensed milk','specialty cheese','frozen potato products','finished products','curd cheese','cream','frozen fruits','frozen chicken']
shelves = shelf1 + shelf2 + shelf3 + shelf4 + shelf5 + shelf6 + shelf7
print(len(shelves), 'of them are already placed in the shelves')
#print(shelves)
```

```python
# We calculate the products that are left to place
SKUs_to_place = [x for x in SKUs if (not x in shelves)]
print('We have', len(SKUs_to_place), 'new SKUs to place')
print(SKUs_to_place)
```

```python
# Before we start the analysis, we will manually assign some of the products to a shelf. 
# We will do this when it makes no sense placing them somewhere else, either because they need refrigeration or
#because they would be difficult to find by the costumers if they are not placed with items of the same category. 

add_to_shelf3 = ['chicken', 'turkey', 'meat'] # meat from different animals
                                              # need refrigerationn and to be with similar already placed products like 'pork'
add_to_shelf4 = ['female sanitary products', 'dental care'] # need to be with the hygene and personal care products
add_to_shelf7 = ['UHT-milk', 'butter milk', 'cream cheese', 'frozen fish', 'frozen vegetables', 
                 'hard cheese', 'ice cream', 'margarine', 'processed cheese', 'sliced cheese', 
                 'soft cheese', 'whipped/sour cream'] # dairy and frozen products
                                                      # need refrigeration and to be with others dairy and frozen products
shelf3 = shelf3 + add_to_shelf3
shelf4 = shelf4 + add_to_shelf4
shelf7 = shelf7 + add_to_shelf7
shelves = shelf1 + shelf2 + shelf3 + shelf4 + shelf5 + shelf6 + shelf7
```

```python
# We calculate again the products that are left to place
SKUs_to_place = [x for x in SKUs if (not x in shelves)]
print('We have now', len(SKUs_to_place), 'SKUs to place')
print(SKUs_to_place)
```

```python
# We will now calculate all the possible combinations of two products, one that's already placed and one that we have to place.
# There's no need to calculate the combinations between two products that are already placed since we are not able to move them.
# This way we will have less combinations and it will be quicker to compute the metrics for each of them.

# We exclude shelf 7 from the analysis since it makes no sense putting the products that are left to place toghether with dairy or frozen porducts,
#mainly because they don't need that kind of refrigeration.
# And also items from shelf 7 like some dairy product are among the most sold products, 
#so their high individual support can interfere with the metrics of other products, espicially less sold ones.

relevant_shelves = shelf1 + shelf2 + shelf3 + shelf4 + shelf5 + shelf6
combi = list(it.product(relevant_shelves, SKUs_to_place))
print('Possible combinations of an item already placed and one to be placed:', len(combi))
combi
```

### Analysis

```python
# This is the proffesor's code to calculate the metrics, we just added the formula to also calculate the confidence.
# Since confidence is not simetric like the other metrics, that is, confindence(X->Y) != confindence(Y->X), 
#we decided to compute confindence(X->Y), where X is an item already placed and Y is one of the items to be placed.
# This way a higher confidence means that we have a higher probability of someone purchasing Y given that they purchased X.

support = {}
lift = {}
leverage = {}
confidence = {}
count_baskets = len(baskets)
# Not a very efficient way to perform the analysis, but it works
for c in combi:
    item1 = c[0]
    item2 = c[1]
    item1_count = 0
    item2_count = 0
    item_combi_count = 0
    for x in baskets:
        if item1 in x:
            item1_count += 1
            if item2 in x:
                item_combi_count += 1
        if item2 in x:
            item2_count += 1
    support[c] = item_combi_count/count_baskets
    lift[c] = item_combi_count / item1_count/ item2_count * count_baskets
    leverage[c] = item_combi_count / count_baskets - item1_count * item2_count / count_baskets**2
    confidence[c] = item_combi_count / item1_count
```

We will put the relevant information in a pandas DataFrame for an easy visualization. We are now able to see, for each combination of products, one already placed and one to place, the value of the four diffferent metrics that we will be using to do the analysis. 

```python
support_values = list(support.values())
lift_values = list(lift.values())
leverage_values = list(leverage.values())
confidence_values = list(confidence.values())
items_placed = []
items_to_place = []
for x, y in combi:
    items_placed.append(x)
    items_to_place.append(y)
    
my_dict = {'Items placed': items_placed, 'Items to place': items_to_place, 
           'Support': support_values, 'Lift': lift_values, 'Leverage': leverage_values, 'Confidence': confidence_values}
df = pd.DataFrame(my_dict)
df.head()
```

The idea is now to calculate for each item to place, which is the item placed that maximizes each metric

```python
# Let's do it for one product, 'jam ', to see how it works    
copy = df.copy()
copy = copy[copy['Items to place'] == 'jam']

# We will eliminate some combinations by asking them to have an acceptable lift and leverage and then taking the max in each category,
#otherwise we could end up with a product that has a very high lift, but an unacceptable leverage, for example.
# By acceptable we mean that it shows a positive correlation, even if it is a small one.
# We also eliminate combinations with very low support
copy = copy[(copy['Lift']>1) & (copy['Leverage']>0) & (copy['Support']>0.001)]
df1 = copy[copy['Lift'] == copy['Lift'].max()]
df2 = copy[copy['Leverage'] == copy['Leverage'].max()]
df3 = copy[copy['Confidence'] == copy['Confidence'].max()]
index = ['Combination with best lift for jam', 'Combination with best leverage for jam', 'Combination with best confidence for jam']
results1 = pd.concat([df1,df2,df3], keys=index)
results1
```

From this, we will just store the item already placed that gets us the best value for each metric, so for each product we will end up with something like this:

```python
my_dict2 = {'Item to place': 'jam', 
            'Item with best Lift': df1.iloc[0,0], 
            'Item with best Leverage': df2.iloc[0,0], 
            'Item with best Confidence': df3.iloc[0,0]}
results2 = pd.DataFrame(my_dict2, index=[''])
results2
```

And then we will calculate the product that appears the most [(mode)](https://en.wikipedia.org/wiki/Mode_(statistics)) considering the three columns 'Item with best Lift', 'Item with best Leverage', 'Item with best Confidence'. In this case, most of the metrics agree that 'beef' is the best product to place 'jam' with, so we will then add 'jam' to shelf 3 (which makes sense since in that shelf there are also 'syrup' and 'honey' that have similar uses). We can still consider 'root vegetables' as a second option in case we need that for another product.

Let's do the same analysis but for all products at once.

```python
# For all the products
best_lift = []
best_leverage = []
best_confidence = []
for product in SKUs_to_place:
    copy = df.copy()
    copy = copy[copy['Items to place'] == product]
    copy = copy[(copy['Lift']>1) & (copy['Leverage']>0) & (copy['Support']>0.001)]
    if copy.empty:
        # for some products not frequently sold, asking support > 0.001 left us with no combinations, 
        #so in that case we ask for less support.
        copy = df.copy()
        copy = copy[copy['Items to place'] == product]
        copy = copy[(copy['Lift']>1) & (copy['Leverage']>0) & (copy['Support']>0.0001)]
    df1 = copy[copy['Lift'] == copy['Lift'].max()]
    df2 = copy[copy['Leverage'] == copy['Leverage'].max()]
    df3 = copy[copy['Confidence'] == copy['Confidence'].max()]
    best_lift.append(df1.iloc[0,0])
    best_leverage.append(df2.iloc[0,0])
    best_confidence.append(df3.iloc[0,0])

my_dict3 = {'Item to place': SKUs_to_place, 
            'Item with best Lift': best_lift, 
            'Item with best Leverage': best_leverage, 
            'Item with best Confidence': best_confidence}
final = pd.DataFrame(my_dict3)
results = final.mode(axis=1)

final['We should put it with'] = results
final
```

With the information of the 'We should put it with' column we can now assign each product to a shelf.

```python
# We define a function that given an already placed product and the products in each shelf returns in which shelf that product is.
def is_in_shelf(product, shelves):
    l = []
    for shelf in shelves:
        if product in shelf:
            l.append(True)
        else:
            l.append(False)
    return l.index(True)+1

# Example
s = [shelf1, shelf2, shelf3, shelf4, shelf5, shelf6, shelf7]
product = 'whole milk'
shelf = is_in_shelf(product, s)
print(product, 'is in the shelf number', shelf)
    
```

```python
# Now I assign the products according to the 'We should put it with' column
s = [shelf1, shelf2, shelf3, shelf4, shelf5, shelf6, shelf7]
add_to_shelves = {'1': [], '2': [], '3': [], '4': [], '5': [], '6': [], '7': []}
for i, product in enumerate(SKUs_to_place):
    with_product = final.iloc[i,-1]
    n = is_in_shelf(with_product, s)
    add_to_shelves[str(n)].append(product)

for i in range(6):
    print("To add in shelf " + str(i+1) + ": " + str(len(add_to_shelves[str(i+1)])))
    print(add_to_shelves[str(i+1)])
    print("\n")
```

We should still check if this distribution makes sense for us taking into account our retail knowledge. And if we are not convinced about the placing of one of the products, we can go for the second option (product that didn't appear the most considering the three columns), which is still good according to one of the metrics, since it was the best in that category. For example for 'mayonnaise', the first option was to put it with chocolates and candies, but the second option is to put it with 'other vegetables' since that is the product that maximizes the leverage for 'mayonnaise'. We like that option better since then it will be in the same shelf than 'ketchup' and 'sauces' that have more similar uses.

We can also place a product in both the first and second options. For example, for 'onions' the algorithm tells us to put it in the shelf where all the meat is. This may be becuase people usually buy them together, the problem is that if someone doesn't go to the meat shelf, they won't find the onions, since the logical place to find them is with the other vegetables (the second option). So by putting a portion in each shelf we can be sure that people will find it and also be encouraged to buy it together with meat.

So, after choosing the second option for the following items: 'mayonnaise', 'misc. beverages', 'beverages' and placing 'onions' in the two options, the final distributions of the shelves is:

```python
final_shelves = []
for i, shelf in enumerate(s):
    shelf = shelf + add_to_shelves[str(i+1)]
    final_shelves.append(shelf)
final_shelves[4].remove('mayonnaise'), final_shelves[4].remove('misc. beverages'), final_shelves[4].remove('beverages')
final_shelves[5].append('onions'), final_shelves[5].append('mayonnaise'), final_shelves[1].append('misc. beverages'), final_shelves[1].append('beverages')

for i in range(7):
    print("Final shelf " + str(i+1) + ": " + str(len(final_shelves[i])))
    print(final_shelves[i])
    print("\n")
```
